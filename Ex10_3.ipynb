```python
import numpy as np
import matplotlib.pyplot as plt
x =np.linspace(0, 10, 500)
plt.plot(x,np.exp(x/-10)*np.sin(np.pi*x))
plt.plot(x,np.exp(x/-3)*x)
plt.xlabel ('Data')
plt.ylabel ('f & g functions value')
plt.legend (['func f', 'func g'])
```

```python
t = np.linspace(0, 2, 100)
theta = t*np.pi
r0 = 0.8
r = r0 + np.cos(theta)
x = r*np.cos(theta)
y = r*np.sin(theta)
plt.plot(x,y)

r0 = 1
r = r0 + np.cos(theta)
x = r*np.cos(theta)
y = r*np.sin(theta)
plt.plot(x,y)

r0 = 1.2
r = r0 + np.cos(theta)
x = r*np.cos(theta)
y = r*np.sin(theta)
plt.plot(x,y)

plt.legend(['r0 = 0.8', 'r0 = 1', 'r0 = 1.2'])
```
